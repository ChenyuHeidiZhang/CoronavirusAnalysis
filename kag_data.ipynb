```python
def get_max_confirmed(dataset):
    return dataset.max(axis=0)[5]
def plot_confirmed_over_provice(main_set, country_names):
    nums = list()
    datasets = list()
    for name in country_names:
        datasets.append(main_set[main_set['Country/Region'] == name])
    for data in datasets:
        max_num = get_max_confirmed(data)
        nums.append(max_num)
    print(nums) 
    plt.scatter(country_names, nums)
```

Now work on Time-Serie Data

```python
def clean_time_data(data):
    data = data.drop('Lat', axis=1).drop("Long", axis=1)
    data = data.dropna(axis=1)
    data = data.groupby('Country/Region').sum()
    return data
```

```python
confirm = clean_time_data(pd.read_csv(directory + 'time_series_covid_19_confirmed.csv'))
death = clean_time_data(pd.read_csv(directory + 'time_series_covid_19_deaths.csv'))
recovered = clean_time_data(pd.read_csv(directory + 'time_series_covid_19_recovered.csv'))
confirm.head()
```

```python
China_confirm = confirm.loc['Mainland China']
China_death = death.loc['Mainland China']
China_recovered = recovered.loc['Mainland China']
plt.plot(China_confirm)
plt.plot(China_death)
plt.plot(China_recovered)
plt.legend(['confirm', 'death', 'recovered'])
# US_confirm = confirm.loc['US']
# US_confirm
```

## Define RNN Model

```python
from sklearn.preprocessing import MinMaxScaler

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
length = len(China_confirm)
all_data = np.empty((length, 3))
for i in range(length):
    all_data[i][0] = China_confirm[i]
for i in range(length):
    all_data[i][1] = China_death[i]
for i in range(length):
    all_data[i][2] = China_recovered[i]

sc = MinMaxScaler(feature_range=(0,1))
all_data_scale = sc.fit_transform(all_data)
```

```python
train_len = 6*length//7
X_train = [] #shape 26,3,10 timestep = 26, channel/feature = 3, series used to predict = 10
Y_train = []
for i in range(10, train_len):
    X_train.append(all_data_scale[i-10:i,:])
    Y_train.append(all_data_scale[i,:])
X_train, Y_train = np.array(X_train), np.array(Y_train)
X_train.shape
```

```python
X_test = [] #shape 26,3,10 timestep = 26, channel/feature = 3, series used to predict = 10
Y_test = []
for i in range(train_len, length):
    X_test.append(all_data_scale[i-10:i,:])
    Y_test.append(all_data_scale[i,:])
X_test, Y_test = np.array(X_test), np.array(Y_test)
Y_test.shape
```

```python
def RNN(X_train, Y_train):
    # Build The Model Now:
    regressor = Sequential()
    # Adding the first LSTM layer and some Dropout regularisation
    regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 3)))
    regressor.add(Dropout(0.2))

    # Adding a second LSTM layer and some Dropout regularisation
    regressor.add(LSTM(units = 50, return_sequences = True))
    regressor.add(Dropout(0.2))

    # Adding a third LSTM layer and some Dropout regularisation
    regressor.add(LSTM(units = 50, return_sequences = True))
    regressor.add(Dropout(0.2))

    # Adding a fourth LSTM layer and some Dropout regularisation
    regressor.add(LSTM(units = 50))
    regressor.add(Dropout(0.2))

    # Adding the output layer
    regressor.add(Dense(units = 3))
    # Compiling the RNN
    regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

    # Fitting the RNN to the Training set
    regressor.fit(X_train, Y_train, epochs = 10, batch_size = 8)
    return regressor
```

```python
model = RNN(X_train, Y_train)
```

```python
# convert back and check prediction
predict = model.predict(X_test)
predict = sc.inverse_transform(predict)
predict.shape
```
